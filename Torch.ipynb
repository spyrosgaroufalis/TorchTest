**TORCHN.NN ( NEURAL NETWORK) CONTAINS EVERYTHING WE NEED**

**LINEAR REGRESSION formula= Yi= f(Xi,b)+ei , GENERAL FORM IS Y=a+bX -> X IS EXPLANATORY VAR AND Y IS DEPENDENT VAR**

```python
import torch
from torch import nn
import matplotlib.pyplot as plt

# check
torch.__version__
```

```python
# creating known parameters
weight = 0.7 # equivalent to b
bias = 0.3 # equivalent to a

start = 0
end = 1
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1) # input
y = weight*X+bias # output

X[:10], y[:10]
```

**TRAINING-VALIDATING-TESTING SETS**

**70% - 15 % - 15% OR 80-20 MOSTLY**

```python
len(X), len(y)
```

```python
# creating training, test split
train_split = int(0.8*len(X)) # this is 40 (80% of lenX which is 50)
X_train, y_train = X[:train_split], y[:train_split]
X_test, y_test = X[train_split:], y[train_split:]

len(X_train), len(y_train), len(X_test), len(y_test)
```

```python
# X_train, y_train

# how to visualise it?
# simple function plots train data and test data and compares predictions
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=None):

  plt.figure(figsize=(10,7))

  # plot training data in blue
  plt.scatter(train_data, train_labels, c="b", s=4, label="Training data")

  # plot test data in green
  plt.scatter(test_data, test_labels, c="g", s=4, label= "Testing data")

  # if predictions?
  if predictions is not None:
            # PLOT them
            plt.scatter(test_data, predictions, c="r", s=4, label="Predictions")

plt.legend(prop={"size": 14});
```

```python
plot_predictions();
# no predictions so no red dots
```

```python
#Building the model with linear regression
```

```python
from torch import nn

class LinearRegressionModel(nn.Module): # it inherits from nn.Module (subclasses it)
  def __init__(self):
    super().__init__()
    self.weights = nn.Parameter(torch.randn(1, ## weight, bias are model parameters
                                            requires_grad=True,
                                            dtype=torch.float))
    self.bias = nn.Parameter(torch.randn(1,
                                          requires_grad=True,
                                          dtype=torch.float))

  # forward defines the computation in the model
  # any subclass of nn.module needs to override forward method
  def forward(self, x: torch.Tensor) -> torch.Tensor: # x is input data
    return self.weights * x + self.bias # LINEAR REGR FORMULA

# it takes random value weight, bias and tries to recreate the pattern by updating values
# it does that by looking the training data and adjusting to get closer to ideal values
# it does that with 2 main algorithms: GRADIENT DECENT, BACKPROPAGATION

```

```python
# how to check whats going on?
# create random seed
torch.manual_seed(42)

# create instance of the mdoel
model_0 = LinearRegressionModel()

#check out the params
list(model_0.parameters())
```

```python
# list named paramas
model_0.state_dict()
```

```python
# making predictions
# using torch.inference_mode()
# how well it predicts y_test based on X_test

with torch.inference_mode(): # this removes the gradient info that is needed in gradient descent
  y_preds = model_0(X_test)

y_preds

# the same is: with torch.no_grad(): but there are some minor diff
```

```python
# y_test
plot_predictions(predictions=y_preds)
# apoti blepoume einai makria apta ideal data
```

```python
list(model_0.parameters())
```

**we need : training loop, testing loop**

**THERE ARE MULTIPLE LOSS FUNCTIONS**

**THE LEARNING RATE BASICALLY SAYS WITH WHAT RATE (STEP) TO ADJUST THE PARAMETERS TO OPTIMISE**

**GOAL= REDUCE MAE = MEAN ABSOLUTE ERROR**

```python
# setup loss fun
loss_fn = nn.L1Loss()

# set up an optimiser
# there are multiple algorithms ( 1 IS SGD stochastic gradient descent)
optimizer = torch.optim.SGD(params=model_0.parameters(),
                            lr= 0.01) # learning rate -> the mos timportant hyperparameter
# sgd adjusts parameters to reduce the loss
```

**GRADIENT= SLOPE OF AN EUTHEIA(klish)**

```python
epochs = 50 # epoch is 1 loop through the data ( hyperparameter)

# 1) loop through data
for epoch in range(epochs):
  model_0.train() # train mode in pytorch sets parameters that require gradients to require gradients

  # 2) forward pass
  y_pred = model_0(X_train)

  # 3) calculate loss
  loss = loss_fn(y_pred, y_train)

  # 4) optimizer zero grad
  optimizer.zero_grad()

  # 5) loss backward
  loss.backward() # compute the gradient of every param with req_grad=true

  # 6) optimizer step( performs gradient descent)
  optimizer.step() # if we didnt zero it in 4) in the loops it will accumulate

  # print out what's happening
  print(f"epoch: {epoch} loss: {loss}")

  model_0.eval() # turn off gradient tracking
  # testing
  with torch.inference_mode(): # turn off grad tracking
    test_pred = model_0(X_test)# do forward pass
    test_loss = loss_fn(test_pred, y_test) # calc the loss

  print(f"test loss: {test_loss}")
```
