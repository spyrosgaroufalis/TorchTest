```python
steps = torch.arange(start=0,end=1000, step=77)
torch.arange(0,10), steps
```

TENSOR DATATYPE
TENSOR SHAPE
TENSOR DEVICE

```python
more_zeros = torch.zeros_like(input= steps,
                              dtype=torch.float16, # tensor datatype(32 is default)
                              device=None, # "cuda" , what dev is your tensor
                              requires_grad=False) # if you want torch to track gradients of a tensor when it goes to sum calculations
# more_zeros
# more_zeros.dtype
# change type
more_zeros16ver = more_zeros.type(torch.float32)
# more_zeros16ver.dtype
print(more_zeros16ver)
print(f"type: {more_zeros16ver.dtype}")
print(f"shape: {more_zeros16ver.shape}") # h to function size(), to shape einai attribute
print(f"device: {more_zeros16ver.device}")
```
